```python
1.activeCount() : This function returns the number of currently active Thread objects in the current thread's thread control block.
                  It gives you the count of threads that are currently running or in a runnable state. This function is useful when
                   you want to monitor the number of active threads in your program.
        
2.currentThread() : This function returns a reference to the currently executing Thread object. It allows you to obtain the instance
                    of the Thread object representing the thread in which the function is called. You can use this function to access
                    properties and methods of the current thread, such as its name or identifier.

3.enumerate() :  This function returns a list of all currently active Thread objects. It provides a way to obtain a list of all running
                threads in the current program. The returned list contains Thread objects that represent the currently running threads.
                This function is useful when you need to iterate over all active threads and perform some operations on them, such as 
                 checking their status or terminating them.
```

```python
3. Explain the following functions

1. run() : The run() method is the entry point for the thread's execution. It contains the code that will be executed in
          a separate thread when the start() method is called. You can override this method in a subclass of the Thread
           class to define the specific behavior of the thread. The run() method encapsulates the task or instructions 
           that you want the thread to perform.    
    
2. start() : The start() method is used to start the execution of a thread. It creates a new thread of execution and calls
              the run() method within that thread. Once the start() method is called, the thread begins executing concurrently
              with other threads. It is important to note that the run() method is automatically invoked when start() is called,
              and you should not call run() directly. Calling start() allows the thread to run in parallel, while calling run() 
               directly will make it run in the same thread as the caller.
    
3. join() :  The join() method is used to wait for a thread to complete its execution. When you call join() on a thread object,
             the calling thread (typically the main thread) will pause its execution until the specified thread has finished executing. 
             This is useful when you want to ensure that the main thread waits for other threads to complete their tasks before continuing.
             By using join(), you can synchronize the execution of multiple threads.
    
4. isAlive() : The isAlive() method is used to check whether a thread is currently running or alive. It returns a boolean value (True or False)
               indicating the thread's status. If the thread is still running, the method will return True; otherwise, it will return False.
               This method allows you to dynamically check the status of a thread and take appropriate actions based on its current state.
```

```python
Q4. Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes
```

```python
import threading

def square():
    sq = [num**2 for num in range(1,11)]
    for i in sq:
        print(i)
        
def cube():
    cb = [num**2 for num in range(1,11)]
    for i in cb:
        print(i)
        
thread1 = threading.Thread(target=square)
thread2 = threading.Thread(target=cube)

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print("finished the executing the program")
```

```python
Q5. State advantages and disadvantages of multithreading

Ans: Advantage of multithreading
   
    1. Doing Multiple Things at Once: Multithreading allows your program to do multiple things simultaneously.
       It can make your program faster and more efficient by using the available resources better.
        
    2.Responsive Program: With multithreading, your program can remain responsive even when it is doing time-consuming 
       tasks. It prevents the program from freezing or becoming unresponsive to user actions.
        
    3.Efficient Resource Usage: Multithreading helps make better use of your computer's resources, like the CPU and
      memory. By running multiple threads at the same time, you can utilize the available processing power more effectively.
        
    4. Faster Input/Output (I/O) Operations: Multithreading is useful for programs that involve reading from or writing to files,
       network communication, or working with databases. While one thread is waiting for I/O to complete, other threads can continue 
        executing, making the overall process faster.
        
    5. Parallelism for Certain Tasks: Although there are limitations, multithreading can offer parallelism for certain tasks that don't
      require the Global Interpreter Lock (GIL). This means tasks like mathematical computations or data processing can take advantage of 
     multiple CPU cores and run faster.

     
    Disadvantage of Multithreading :
        
    1.Python's Global Interpreter Lock (GIL): Python has a limitation called the GIL that prevents true parallel execution of threads. 
      This means that even with multithreading, only one thread can execute Python code at a time. It limits the potential performance 
      gains of multithreading, especially for tasks that are CPU-bound.
    
    2. Increased Complexity: Multithreading adds complexity to your program. You need to carefully manage synchronization between threads 
       to avoid issues like race conditions or deadlocks. Debugging and understanding multithreaded code can be more challenging due to the 
        potential for unexpected interactions between threads.
        
    3. Additional Overhead: Multithreading has some overhead due to the creation and management of threads, as well as the synchronization
       mechanisms used to coordinate their execution. These additional tasks can impact performance and make the program slower.
        
    4.Bugs and Difficult Debugging: Multithreading can introduce bugs that are difficult to reproduce and fix. Race conditions, where the 
      behavior of the program depends on the timing of thread execution, can be particularly challenging to identify and resolve.
        
    5. Limited Scalability: Because of the GIL, multithreading in Python may not scale as well as other languages or threading models that
       allow true parallelism. If your program requires high scalability, alternative approaches like multiprocessing or asynchronous programming 
        might be more suitable.
    
```

```python
Q6. Explain deadlocks and race conditions.

Ans: Deadlock :
       Deadlock refers to a situation where two or more threads or processes are unable to proceed because each is waiting for 
       the other to release a resource or complete a task. As a result, the threads end up in a circular wait, leading to a 
        standstill. Deadlocks can occur in multithreaded or multiprocessing environments when certain conditions are met:
            
    1.  Mutual Exclusion: Resources involved in the deadlock must be non-shareable, meaning only one thread can use a resource at a time.

    2. Hold and Wait: A thread holding at least one resource is waiting to acquire additional resources held by other threads.

    3. No Preemption: Resources cannot be forcefully taken away from threads holding them. They can only be released voluntarily.

    4. Circular Wait: A set of threads is waiting for resources in a circular manner, where one thread is waiting for a resource held
       by another thread in the set.
        
      -  To resolve deadlocks, techniques like resource allocation strategies, careful ordering of resource acquisition, deadlock detection
        algorithms, and deadlock prevention policies can be employed.
        
    
    - Race conditions :
        
        Race conditions occur when the behavior of a program depends on the relative timing or interleaving of multiple threads accessing shared 
        resources or variables. It arises when two or more threads access a shared resource concurrently, and at least one thread modifies the
        resource. The outcome of the program becomes unpredictable and can lead to incorrect results. Race conditions can happen due to the absence 
        of proper synchronization mechanisms in multithreaded programs.
        
        
         - For example, consider two threads attempting to increment a shared variable count:

            Thread 1: count = count + 1
            Thread 2: count = count + 1

        If both threads execute simultaneously, they may read the same initial value of count, perform their computations,
        and then overwrite each other's changes. As a result, the final value of count may not reflect the intended increment.  
        
      - To avoid race conditions, synchronization techniques like locks, mutexes, semaphores, or atomic operations are used to ensure exclusive
        access to shared resources. These mechanisms enable proper coordination and control the access to critical sections of code, preventing
        simultaneous modifications by multiple threads.
        
        
        
```
